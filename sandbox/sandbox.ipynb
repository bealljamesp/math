```python
import sympy as sp

# Define pristine mathematical symbols
x = sp.Symbol("x", real=True)
n1 = sp.Symbol("n1", integer=True, positive=True)
n2 = sp.Symbol("n2", integer=True, positive=True)

# Construct the orthogonal integrand: sin(n1*x) * sin(n2*x)
integrand = sp.sin(n1 * x) * sp.sin(n2 * x)

# Integrate over a full cycle from 0 to 2*pi assuming n1 != n2
orthogonal_integral = sp.integrate(integrand, (x, 0, 2 * sp.pi))
print(f"Integral when n1 != n2: {orthogonal_integral}")  # Outputs: 0

# Now force n1 == n2 (integrating a wave against itself)
same_frequency_integral = sp.integrate(sp.sin(n1 * x) ** 2, (x, 0, 2 * sp.pi))
print(f"Integral when n1 == n2: {same_frequency_integral}")  # Outputs: pi
```
